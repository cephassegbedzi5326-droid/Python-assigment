```python
# Items and their prices
Items_prices={
    "Bread":"25",
    "Milk":"18",
    "Eggs":"2.5",
    "Sugar":"20"
    }
```

```python
quantities={}
```

```python
# ask the user for the quantity of items to buy
for item in Items_prices:
    amount = int(input(f"how many {item} do you want to buy"))
    quantities[item] = amount
```

```python
# calculating the total cost
Total_cost=0
for item in Items_prices:
    price=float(Items_prices[item]) # Convert price to float
    qty=quantities[item]
    Total_cost += price * qty
    print(f"\nInitial total cost:{Total_cost}GHs")
```

```python
if Total_cost >100:
  discount=Total_cost *0.1
  total_cost=Total_cost-discount
  print(f"\nTotal cost after discount:{total_cost}GHs")
else:
  print(f"No discount applied.")

```

```python
# calculating subtotal
subtotal=0
for item in Items_prices:
    price=float(Items_prices[item])
    qty=quantities[item]
    subtotal += price * qty


```

```python
# calculating discount on the subtotal
discount=0
if subtotal >100:
   discount=subtotal*0.1
```

```python
#calculating total
total=subtotal-discount
```

```python
# printing the receipt
print("\n-----Receipt-----")
for item in Items_prices:
    Item_total=float(Items_prices[item]) * quantities[item]
    print(f"{item}(x{quantities[item]}):GHS {Item_total:.2f}")
    print("-------------")
    print(f"Subtotal : GHS {subtotal:.2f}")
    print(f"Discount : GHS {discount:.2f}")
    print(f"Total : GHS {total:.2f}")
```

```python
# List of dictionaries that will store studebt names and scores
students=[
    {"name":"Ama", "Maths":78, "Science":82, "English":74},
    {"name":"Kojo", "Maths":65, "Science":59, "English":72},
    {"name":"Esi", "Maths":90, "Science":88, "English":85}

]
```

```python
#Calculating the average for each student using loop
for student in students:
  maths_score = int(student["Maths"])
  science_score =int(student["Science"])
  english_score = int(student["English"])
  # Calculating average for the current student
  average_score = (maths_score + science_score + english_score) / 3
  # Storing average back into the dictionary for the current student
  student.update({"Average":average_score})
```

```python
#printing student's result
for student in students:
  print(
      f"{student["name"]}-Maths:{student["Maths"]},"
      f"Science:{student["Science"]},English:{student["English"]}"
      f"=> Average:{student["Average"]:.2f}"
  )
```

```python
#Determining the student with highest average score
highest_average=0
for student in students:
  if student["Average"] > highest_average:
    highest_average=student
  print("\nStudent with the highest average score:")
  print(f"{highest_average["name"] }= {highest_average["Average"]:.2f}")
```

```python
# Creating a dictionary mapping names to salary
employee_salaries = {
    "Kwame":3200,
    "Akua": 4000,
    "Yaw":2800,
    "Afia":5000
}
```

```python
# Store original salaries for comparison in the next cell
original_salaries = employee_salaries.copy()

# Creating the conditions for salary adjustments and updating employee_salaries
for employee, salary in employee_salaries.items():
  if salary < 3500:
    # Apply 7% increase and update the salary
    employee_salaries[employee] = salary * 1.07
  else:
    # Apply 4% increase and update the salary
    employee_salaries[employee] = salary * 1.04
```

```python
print(f"Kwame | Old:{original_salaries['Kwame']:.2f} | New:{employee_salaries['Kwame']:.2f}")
print(f"Akua | Old:{original_salaries['Akua']:.2f} | New:{employee_salaries['Akua']:.2f}")
print(f"Yaw | Old:{original_salaries['Yaw']:.2f} | New:{employee_salaries['Yaw']:.2f}")
print(f"Afia | Old:{original_salaries['Afia']:.2f} | New:{employee_salaries['Afia']:.2f}")
```

```python
#Creating list of even numbers within the range 2 - 20(inclusive)
even_numbers=list(range(2,21,2))
```

```python
#append 22 to the list
even_numbers.append(22)
```

```python
#remove the smallest number
even_numbers.remove(min(even_numbers))
```

```python
#calculating the sum of the remaining numbers
total_sum=sum(even_numbers)
```

```python
#We will ask the user to input a sentence
sentence=input("Enter a sentence:")
```

```python
#We create a dictionary to store vowel counts
vowel_counts={
    "a":0,
    "e":0,
    "i":0,
    "o":0,
    "u":0
}
```

```python
#converting the characters to lower case
sentence=sentence.lower()
```

```python
#creating loop through each character for vowels
for character in sentence:
  if character is vowel_counts:
    vowel_counts[character] += 1
```

```python
#printing the vowel counts
print("Vowel counts:")
for vowel, count in vowel_counts.items():
  print(f"{vowel}:{count}")
```
